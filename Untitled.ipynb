## Data import with Pandas

- When you want to plot data in Python (from a CSV, or xlsx) you will need in principal two packages

__Pandas__:

Imports the data from a xlsx or CSV file to your Jupyter Notebook as _dataframe_

__Matplotlib__:

Visualises the data you have extracted from the CSV or xlsx file

```python
import pandas as pd

# Import data with pandas 

df  = pd.read_csv(r"data2.csv", delimiter=";")

# delete unimportant rows

df_deleted_rows = df.drop(df.index[0:31])

# select columns for x and y values

x_data = df_deleted_rows["Column1"]
y_data = df_deleted_rows["Column2"]

# transform string values to numeric values 

x_data_num = pd.to_numeric(x_data)
y_data_num = pd.to_numeric(y_data)

```

```python
# pandas dataframe to numpy arrays
import numpy as np

x_data_array = np.array(x_data_num)

print(x_data_array)

multiplied_variable = x_data_array*0.001


```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(x_data_num, y_data_num)

plt.show()

print("done")
```
